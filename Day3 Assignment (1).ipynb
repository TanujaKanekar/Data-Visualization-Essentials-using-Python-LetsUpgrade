```python
#importing necesaary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the data
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
#Regression plot
sns.relplot(x='timepoint',y='signal',hue='event',data=fmri,aspect=3,kind='line')
```

```python
#regression plot
sns.relplot(data=fmri,x='timepoint',y='signal',hue='region')
```

```python
#boxplot 
sns.boxplot(data=fmri,x='timepoint',y='signal',hue='region')
```
