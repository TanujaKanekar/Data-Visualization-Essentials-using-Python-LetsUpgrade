 #### Line Plots:
 >The line graph is a type of chart used to show information that changes over time. We plot line  using several points    connected by straight lines.  We also call it a line chart. The line graph comprises of two axes known as ‘x’ axis and ‘y’ axis.
The  axis is known as the x-axis.
The vertical axis is known as the y-axis.

#### Question1


```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#input
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
#Plotting the line plot
plt.plot(a,b,'c--v',linewidth=2,markersize=10,markerfacecolor='r')
plt.xlabel('a ranging from 40 to 50')
plt.ylabel('b ranging from 50 to 60')
plt.title("Line Plot")
```

#### Question 2

```python
#inputs
days=[1,2,3,4,5,6,7]
sales_1=[160,150,140,145,175,165,180]
sales_2=[70,90,160,150,140,145,175]
```

```python
#plotting the graph for sales1 and sales2
plt.plot(days,sales_1,'orange',label='sales_1')
plt.plot(days,sales_2,'blue',label='sales_2')
plt.xlabel('Days',color='red')
plt.ylabel('Sales',color='red')
plt.title("Days vs Sales plot",color='red')
plt.legend(loc='lower right')
```
